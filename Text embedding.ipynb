我問 ChatGPT: 寫 python, hugging face 呼叫 BAAI/bge-large-zh-v1.5, 把3段文章轉成 embedding vectors, 用 cosine similarity 計算彼此相似度

```python
from transformers import AutoTokenizer, AutoModel
```

```python
import torch
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# 載入模型和 tokenizer
model_name = "BAAI/bge-large-zh-v1.5"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
# 將段落轉換為 embeddings
def get_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
    # 使用模型的輸出向量 (通常為最後一層的隱藏層) 的 [CLS] 標記來表示整體 embedding
    embedding = outputs.last_hidden_state[:, 0, :]
    return embedding.squeeze().numpy()
```

```python
doc1 = "元智大學攜手IC晶圓測試廠欣銓科技於112學年成立半導體書院，雙方展開積極合作，強化半導體產業人才培育與產業鏈結"
doc2 = "元智大學自111學年起成立企業書院，攜手產業和業師共同育才，讓學生認識公司理念與文化、實際了解企業工作環境，提高學生對企業就職的興趣與意願。"
doc3 = "智能康養機器人的AI大腦透過安裝在自走平台上的螢幕與長輩互動，播放聲音與動畫來激勵長輩運動."
```

```python
embedding1 = get_embedding(doc1)
```

```python
print(embedding1)
```

```python
embedding2 = get_embedding(doc2)
embedding3 = get_embedding(doc3)
```

```python
# 計算每對 embeddings 之間的餘弦相似度
similarity_matrix = cosine_similarity([embedding1, embedding2, embedding3])
```
