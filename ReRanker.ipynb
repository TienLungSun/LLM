```python
from FlagEmbedding import FlagReranker
```

```python
reranker = FlagReranker('BAAI/bge-reranker-large', use_fp16=True) # Setting use_fp16 to True speeds up computation with a slight performance degradation
```

```python
doc1 = "元智大學攜手IC晶圓測試廠欣銓科技於112學年成立半導體書院，雙方展開積極合作，強化半導體產業人才培育與產業鏈結"
```

```python
doc2 = "元智大學自111學年起成立企業書院，攜手產業和業師共同育才，讓學生認識公司理念與文化、實際了解企業工作環境，提高學生對企業就職的興趣與意願。"
```

```python
score = reranker.compute_score([doc1,doc2])
print(score)
```

```python
doc3 = "智能康養機器人的AI大腦透過安裝在自走平台上的螢幕與長輩互動，播放聲音與動畫來激勵長輩運動."
score = reranker.compute_score([doc1,doc3])
print(score)
```
